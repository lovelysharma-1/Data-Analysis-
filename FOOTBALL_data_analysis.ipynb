```python
player_salary = fifa[['short_name', 'wage_eur']]
```

```python
player_salary.head()
```

```python
player_salary = player_salary.sort_values(by=['wage_eur'], ascending= False)
```

```python
player_salary.head()
```

```python
plt.figure(figsize=(7,5))
plt.bar(list(player_salary['short_name'])[0:5], list(player_salary['wage_eur'])[0:5], color= ['g'])
plt.title('Top 5 players with highest wages')
plt.xlabel("Player's Name")
plt.ylabel('Wages in euro')
plt.xticks(rotation=45)
plt.show()
```
