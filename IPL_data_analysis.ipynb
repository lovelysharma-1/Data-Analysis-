```python
winner_counts = ipl['winner'].value_counts()
```

```python
top_winners = winner_counts.head(5)
top_winners
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=top_winners.index, y=top_winners.values, palette="viridis")
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.title('Title: Top 5 Winners')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Create a bar plot using matplotlib.pyplot
# Top 5 winners
plt.figure(figsize=(8, 8))
plt.bar(top_winners.index, top_winners.values, color='g')

# Write the number of winning matches above each bar
for index, value in enumerate(top_winners.values):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=12)

plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.title('Title: Top 5 Winner Teams')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()






```

```python
#Extracting the records where a team won batting first
batting_first_winner = ipl[ipl['win_by_runs']!=0]
batting_first_winner
```

```python
#Making a histogram
plt.figure(figsize=(5,5))
plt.hist(batting_first_winner['win_by_runs'])
plt.title('Title: Distribution of Runs')
plt.xlabel('Runs')
plt.ylabel('No.of Matches')
plt.show()
```

```python
# Finding out the number of wins w.r.t. each team after batting first
batting_first_winner['winner'].value_counts()
```

```python
plt.figure(figsize=(10,5))
plt.hist(batting_first_winner['winner'])
plt.title('Title: Batting first winners')
plt.xlabel('Teams')
plt.ylabel('No. of matches')
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python
ipl.shape
```

```python
# Making a bar-plot for top 3 teams with most wins after batting first
plt.figure(figsize=(5,5))
plt.bar(list(batting_first_winner['winner'].value_counts()[0:3].keys()),list(batting_first_winner['winner'].value_counts()[0:3]),color=('blue','yellow','orange'))
plt.xlabel('Teams')
plt.ylabel('No. of matches')
plt.xticks(rotation=45)
plt.show()
```

```python
labels=batting_first_winner['winner'].value_counts().keys()
```

```python
# Making a pie chart
plt.figure(figsize=(10,7))
plt.pie(list(batting_first_winner['winner'].value_counts()),labels=(batting_first_winner['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.title('Title: Winning percentange of teams after batting first')
plt.show()
```

```python
# Extracting those records where a team won after batting second
batting_second_winner= ipl[ipl['win_by_wickets']!=0]
batting_second_winner
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_second_winner['win_by_wickets'],bins=30)
plt.title('Title: Wicket Distribution')
plt.xlabel('Wickets')
plt.ylabel('No. of matches')
plt.show()
```
